```python
import pandas as pd
```

```python
data = pd.read_csv(r"sort_data.csv")
data.head()
```

```python
# i. Add a column named as “City” which store city names extracted from “Purchase Address”
# column. (Hint: use split method on column level)
data['City'] = data['Purchase Address'].str.split(',').str[1]
data.head()
```

```python
# ii. Add a column named as “State” which stores state names extracted from “Purchase Address” column.
data['State'] = data['Purchase Address'].str.split(',').str[2]
data.head()
```

```python
data['State'] = data['State'].str.split(' ').str[1].str.strip('[]')
data.head()
```

```python
# iii. Add a column names as “Zip code” which stores zip codes extracted from “Purchase Address”
# column.
data['Zip code'] = data['Purchase Address'].str.split(',').str[2].str.split(' ').str[2]
data.head()
```

```python
# iv. Remove the column “Purchase Address” from the dataframe.
data.drop(data[['Purchase Address']],axis = 1,inplace = True)
data.head()
```

```python
data.isnull().sum() # due to some above changes the dataframe attain some Nan values so we have to remove it.
```

```python
data.dropna(inplace = True)
data.isnull().sum()
```

```python
# v. Which city made the least sale? Show the name of city along with sales.
city_sales_total = data.groupby('City')['Sales'].sum()

least_city = city_sales_total.idxmin()
print(least_city,':',city_sales_total[least_city])
```

```python
# vi. Add a column named as “Hour” which store the hour in which order was made using the “Order 
# Date” column.
data['Order Date'] = pd.to_datetime(data['Order Date'], errors = 'coerce')
data['Hour'] = data['Order Date'].dt.hour
data.head()
```

```python
# vii. Add a column named as “Minute” which store the minutes in which order was made using the 
# “Order Date” column.

data['Minute'] = data['Order Date'].dt.minute
data.head()
```

```python
# viii. Which hour is best for running advertisements?
hourly_sales = data.groupby('Hour')['Sales'].mean()
hourly_sales
```

```python
best_hour = hourly_sales.idxmax()
print(f"The best hour for running advertisements is {best_hour}:00.")
```

```python
# ix. What is maximum number of products sold in a single order? (Hint: Products sold under same 
# order_id were part of that one order).

orders = data.groupby('Order ID')['Product'].count()
orders.unique() 
```

```python
max_order = orders.max()
print(f"The maximum number of products sold in a single order is {max_order}.")
```

```python
# x. Generate a list of top 5 most sold products.

product_sales = data.groupby('Product')['Quantity Ordered'].sum()
product_sales
```

```python
top_products = product_sales.sort_values(ascending=False).head(5)

print("Top 5 most sold products:")
print(" \n",top_products)
```
