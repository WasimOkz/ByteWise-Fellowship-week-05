```python
#1. Merge all the data into single file and store that file as “all_months_data.csv”. (Hint: use os library)

cwd = os.getcwd()

files  = [file for file in os.listdir(cwd) if file.endswith('.csv')]

data = pd.concat([pd.read_csv(file) for file in files])

data.to_csv('all_months_data.csv')
```

```python
# 3. Check how many nan values are present in each column.

data.isna().sum()
```

```python
# 4. Remove the nan values from the columns.

data.dropna(inplace = True)
data.isna().sum()
```

```python
# 5. Add a column named “Month” which store the respective month of “Order Data”.
data.dtypes
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], errors = 'coerce')
data.dtypes

# here errors = 'coerce' is used to handle error in the datetime col.
```

```python
data['Month'] = data['Order Date'].dt.month
data
```

```python
# 'Price Each' value in the col so we will replace it with 0. it is str which not convertable into float
data['Price Each'].unique()
```

```python
data['Price Each'].replace('Price Each',0, inplace = True)
data['Price Each'].unique()
```

```python
# same for Quantity col

data['Quantity Ordered'].unique()
```

```python
data['Quantity Ordered'].replace('Quantity Ordered', 0, inplace = True)
data['Quantity Ordered'].unique()
```

```python
# 6. Add a column named “Sales” which store the total bill of each order. (Hint: total bill = price * 
# Quantity)


data['Price Each'] = data['Price Each'].astype(float)
data['Quantity Ordered'] = data['Quantity Ordered'].astype(float)
data['Sales'] = data['Price Each']* data['Quantity Ordered']
data.head()
```

```python
# 7. How many sales were made each month? Sort the list result in descending order.
sort_data = data.sort_values(by=['Sales'], ascending=False)
sort_data
```
