```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
#code start here
sns.histplot(data = titanic , x= 'fare'  , color = 'red' )
plt.xlabel('fare')
plt.ylabel('Count')
plt.show()
#code ends here
```

```python
#code starts here
from matplotlib.pyplot import figure
figure(figsize= (10,6) )
sns.boxplot(data = titanic, x = 'class'  , y='age', width = 0.5)
#code ends here
```

In box plot we generate three boxes along x-axis we have class col and along y-axis we have age col so it tells us the value of age in each class

```python
#code starts here
cor = titanic.corr()
figure(figsize= (8,6) )
sns.heatmap(cor)
plt.title('titanic.corr()')
#code ends here
```

heatmap tells us the relation of each one col with others where the value is 1 there the color is white and we can see this relation using the side scale bar
